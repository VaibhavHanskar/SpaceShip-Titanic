# SpaceShip Titanic DataSet
Aim of Dataset is to predict whether a passenger was transported to an alternate dimension during the Spaceship Titanic's collision with the spacetime anomaly.

You can download the Dataset from the given link below

https://www.kaggle.com/competitions/spaceship-titanic/data?select=train.csv


# Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas import Series,DataFrame
import seaborn as sns
```

```python
# To see all the columns present in the dataset
pd.set_option('display.max_columns',None)
```

```python
# Import the train dataset
train_data = pd.read_csv('train.csv')
train_data.head()
```

1. PassengerId - A unique Id for each passenger. Each Id takes the form gggg_pp where gggg indicates a group the passenger is travelling with and pp is their number within the group. People in a group are often family members, but not always.
2. HomePlanet - The planet the passenger departed from, typically their planet of permanent residence.
3. CryoSleep - Indicates whether the passenger elected to be put into suspended animation for the duration of the voyage. Passengers in cryosleep are confined to their cabins.
4. Cabin - The cabin number where the passenger is staying. Takes the form deck/num/side, where side can be either P for Port or S for Starboard.
5. Destination - The planet the passenger will be debarking to.
6. Age - The age of the passenger.
7. VIP - Whether the passenger has paid for special VIP service during the voyage.
8. RoomService, FoodCourt, ShoppingMall, Spa, VRDeck - Amount the passenger has billed at each of the Spaceship Titanic's many luxury amenities.
9. Name - The first and last names of the passenger.
10. Transported - Whether the passenger was transported to another dimension. This is the target, the column you are trying to predict.

# NaN values
Check which all features has NaN values

```python
nan_value = [value for value in train_data.columns if train_data[value].isnull().sum()>0]
nan_value
```

```python
for value in nan_value:
    print("{} has {} NaN values".format(value,train_data[value].isnull().sum()))
```

```python
# Relation between NaN values and Transported Feature
# How NaN values are effecting the Transported Feature

for value in nan_value:
    train_df = train_data.copy()
    
    train_df[value] = np.where(train_df[value].isnull(),"Null","Not Null")
    
    train_df.groupby(value)['Transported'].median().plot.bar()
    plt.xlabel(value)
    plt.ylabel('No. of Transported Passenger')
    plt.title(value)
    plt.show()
```

By looking at the relation between NaN and the Transported Value we can see that in some cases NaN values are effecting our results and in Some cases NaN values are not effecting our results

# Numerical values

```python
numerical_value = [value for value in train_data.columns if train_data[value].dtypes != "O"  and value not in ['PassengerId','Transported']]
numerical_value
```

```python
for value in numerical_value:
    plt.hist(train_data[value],rwidth = 0.95)
    plt.title(value)
    plt.xlabel(value)
    plt.ylabel('Number Of Passenger')
    plt.show()
```

By looking at the Graphs we can see that there are many passengers who has not paid any amount at Luxury Aminities
1. The age columns is almost Normally Distributed
2. Rest other numerical columns are Right skewed then tere must be outliers in the data

```python
for value in numerical_value:
    train = train_data.copy()
    train[value] = np.log(train[value])
    train.boxplot(column = value)
    plt.title(value)
    plt.show()
```
